```python
from classifier import load_data, classify, test

import numpy as np
import matplotlib.pyplot as plt 
```

```python
sizes = np.arange(1, 15, 1)
```

```python
# RF-testing
accuracies = []
F2_scores = []
AUC_scores = []
times = []
for size in sizes:
    X_train, X_test, y_train, y_test = load_data(0.4, size*100000)
    cf, t = classify("rf", X_train, y_train)
    times.append(t)
    acc = test(cf, X_test, y_test, "acc")
    F2 = test(cf, X_test, y_test, "f2")
    AUC = test(cf, X_test, y_test, "auc")
    accuracies.append(acc)
    F2_scores.append(F2)
    AUC_scores.append(AUC)
```

```python
fig, ax1 = plt.subplots()
ax1.plot(sizes, accuracies, linestyle="dashed", label="Accuracy", color='blue')
ax1.plot(sizes, F2_scores, linestyle="dashed", label="F2 score", color='magenta')
ax1.plot(sizes, AUC_scores, linestyle="dashed", label="AUC score", color='purple')

color = 'tab:blue'
ax1.set_xlabel('size of dataset (1e5)')
ax1.set_ylabel('Scores', color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  

color = 'tab:red'
ax2.set_ylabel('Training time (seconds)', color=color)  
ax2.plot(sizes, times, label="time", color="red")
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout() 
ax1.legend()
ax2.legend()
plt.title("Random Forest Classifier")
plt.show()
```
